```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.arange(16).reshape(4,4)
b = np.fliplr(a)
```

```python
fig ,ax = plt.subplots(1, 2)
```

```python
ax[0].imshow(a)
ax[0].set_axis_off()  #子图的去掉 xticks yticks
ax[0].set_title('11')
ax[1].imshow(b)
ax[1].set_title('22')
ax[1].text(1,4, 'This is left')  #左上角是（0， 0）
ax[0].text(1,4, 'This is right') 
ax[1].set_axis_off()
plt.show()
# fig.savefig('1.png')
```
